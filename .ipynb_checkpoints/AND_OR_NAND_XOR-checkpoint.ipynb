sigmoid 통하면 0~ 1 사이의 값이 결과값으로 계산된다

sigmoid 함수는 
classification의 출력이 0~1 사이의 값을 출력하도록 하는 함수

수치미분 함수는
손실함수의 최소값을 찾으면서 가중치 W와 바이어스 b의 값을 업데이트 해주는 미분코드이다

로직게이트 클래스 정의
생성자에서는 입력데이터와 정답데이터를 입력파라미터로 받고 내부적으로 넘파이의 랜드 함수를 이용하여 가중치와 b값을 0~1사이의 값으로 초기화해주고 있다.

다음은 크로스 엔트로피를 이용하여 손실함수값을 계산해주는 로스 펑크 함수와
이런 손실함수 값을 외부에서 알 수 있도록 디스플레이 해주는 에러벨 함수 정의

다음의 트레인 메소드는 로직게이트 클래스의 가장 중요한 함수
트레인 메서드는 객체를 통하여 호출되고, 내부적으로는 수치미분을 이용해 손실함수의 최소값을 찾고 이 과정에서 가중치와 b를 지속적으로 업데이트 해주는 가장 중요한 메소드이다

마지막으로 프리딕 함수는 임의의 임력데이터를 가지고 미래값을 예측해주는 메소드이다. 

즉 우리가 논리게이트 검증을 위해 정리한 로직데이트 클래스는 이렇게 다섯개의 메서드를 가지고 있다. 

-------------

이런 로직게이트 클래스를 객체로 만들기 위하여 사용하기 위한 사용법은 다섯단계로 나눌 수 있다
먼저 입력, 정답 데이터를 넘파이타입으로 선언
우리가 원하는 논리게이트를 만들기 위해 생성된 입력/정답 데이터를 파라미터로 넣어주고 있다. 우리의 예에서는 엔드게이트 만들기 위해 총 네개 입력데이터에 대하여 엔드에 해당하는 정답 데이터가 생성되어 들어가는 것을ㅇ 볼 수 있다.
이렇게 생성된 로직게이트의 객체를 이용하여 손실함수의 최소값을 찾아주는 트레인 메소드를 찾아 내부적으로 최적값을 찾는 w와 b를 찾아가도록 해준다. 
이렇게 손실함수 최소가 되어 학습을 모두 마쳤다면 최종적으로 우리가 원하는 입력 데이터를 넣어 답 데이터를 출력하는지 검증할 수 있는 프리딕 메서드를 호출해 최종 결과값을 확인한다.



-----------


시그모이드와 수치미분 함수는 
소스코드가 동일한것을 알 수 있다.




------
코드에서 
멤버 변수를 나타내는 x데이터, t데이터, w, b, 학습률이 모두 private으로 선언되어있다. 파이썬 클래스의 멤버변수는 기본적으로 퍼블릭이고 외부에서 쉽게 값을 수정하거나 삭제할 수 있고, 머신러닝에서 중요한 가중치나 b 등은 외부에서 함부로 접근해서 바꾸면 안되기 때문에 모두 private으로 선언했다. 그래서 이 값들에 대해서는 내부의 메서드만을 이용해서 변경할 수 있도록 안전장치를 마련한 것이다. 

-------
트레인 메서드에서는
수치미분을 이용하여 손실함수가 최소가 될 때 까지 학습하는 함수로
가중치 w와 b를 업데이트 해주는 함수이다
가중치와 바이어스를 업데이트 하기 위해 외부함수로 정의한 수치미분을 이용하고 있다.

프리딕 함수 또한 이미 입력데이터에 대해 리니어 리그레션 계산 하는 부분과 이렇게 계산된 리니어 리그레션 결과가 시그모이드에 입력으로 들어가서 0이나 1 값을 리턴해주는 함수임을 알 수 있다.
